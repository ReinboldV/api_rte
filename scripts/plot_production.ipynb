Import packages

```python
tstart = datetime.datetime.today()-datetime.timedelta(days=2)
tend   = datetime.datetime.today()+datetime.timedelta(days=1)

r = get_prod(start_date=tstart, end_date=tend, production_type='WIND', type='D-1')
r_2 = get_prod(start_date=tstart, end_date=tend, production_type='WIND', type='CURRENT')
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('Load factor (%)', color=color)
ax1.tick_params(axis='y', labelcolor=color)
r[0]['values']['load_factor'].plot(fig=fig, ax=ax1, color=color)

ax2 = ax1.twinx()
color = 'tab:blue'
r[0]['values']['value'].plot(ax=ax2)
ax2.set_ylabel('Power (MW)', color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax1.grid(True)
```
